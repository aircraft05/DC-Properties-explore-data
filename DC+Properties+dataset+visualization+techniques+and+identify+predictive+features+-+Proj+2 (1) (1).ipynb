## DC Properties dataset visualization tech & exploration

```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline

```

```python
# READ DATA
location = 'DC_Properties.csv'
df = pd.read_csv(location)
df.head()
```

```python
#set INDEX
df.set_index('Unnamed: 0', inplace=True)
df.head()
```

```python
# SHOWS SHAPE OF DATAFRAME (ROWS AND COLUMNS)
df.shape
```

```python
# DATA TYPES FOR EACH COLUMN
df.dtypes
```

```python
# SHOWS HOUSES SOLD FOR LESS THAN $500,000 WITH GREATER THAN 2 BATHROOMS & ACTUAL YEAR BUILT AFTER 1990
df[(df["PRICE"] <= 500000) & (df["BATHRM"] >= 2) & (df["AYB"] > 1990)]
```

```python
# GET A PICTURE OF THE DATA AND WHAT IS REPRESENTED 
#(SHOWS 2ND, 3RD & 4TH QUARTILES ALONG W/ NO. RECORDS, MEAN, STD DEV, MIN & MAX OF DATA)
df.describe()
```

```python
# SHOWS COLUMN NAMES
df.columns
```

```python
# PRICE OF HOUSES SOLD GROUPED BY ZIP CODE AND SHOWING SOME STATISTICAL DATA
df['PRICE'].groupby(df['ZIPCODE']).describe()
```

```python
#CHECK FOR DUPLICATE DATA TO DETERMINE IF SHOULD BE REMOVED
dupe = df.duplicated() 
df[dupe] 
```

```python
#REMOVING THE DUPLICATE DATA
df = df.drop_duplicates()
```

```python
#CHECK FOR MISSING DATA 
df_no_missing = df.dropna()
df_no_missing
```

```python
# VERIFY THE SHAPE OF THE DATA AFTER REMOVING THE DUPLICATES AND ANY OTHER CHANGES
df.shape   #check shape of dataframe after removing duplicates
```

```python
df.isna()
```

```python
# FILL NA VALUES WITH 0 AND REASSIGN DATAFRAME TO INCLUDE 0 FOR NA VALUES
df = df.fillna(0)
```

```python
# CHECK IF ALL COLUMNS ARE NOW ACCOUNTED FOR AND MATCH 157449 ROWS WITH NON=NULL VALUES
df.info()
```

```python
# FIND VARIABLES WHICH ARE CORRELATED TO PRICE OF HOUSE IN DC
df.corr
```

```python
import statsmodels.formula.api as smf
```

```python
# GOT TIRED OF SEEING WARNING - ONLY WANT TO SEE THEM ONCE TO KNOW IF TO TAKE ACTION ON THEM 
import warnings
warnings.filterwarnings(action='once')
```

```python
# RUN A REGRESSION MODEL FOR THE DEPENDANT VARIABLE PRICE AGAINST INDEPENDENT VARIABLES TO GET 
# BEST R-SQUARED & ADJ R-SQUARED CLOSE TO .9 OR HIGHER IF POSSIBLE
result = smf.ols('PRICE ~  ZIPCODE + ASSESSMENT_NBHD + ASSESSMENT_SUBNBHD +YR_RMDL + LONGITUDE + EYB -1', data=df).fit()
```

```python
result.summary()
```
